```python
import pandas as pd
cereal_df = pd.read_csv('cereal.csv')
cereal_df.head()
```

```python
cereal_df.info()
```

```python
cereal_df.describe()
```

```python
cereal_df.sort_values('sugars',ascending=False).head()
```

```python
cereal_df['sugar_per_cup']=cereal_df['sugars']/cereal_df['cups']
```

```python
cereal_df.sort_values('sugar_per_cup',ascending=False).head()
```

```python
cereal_df.sort_values('sugar_per_cup',ascending=False).head(7)
```

```python
cereal_df['protein_cal']=cereal_df['protein']*4
```

```python
cereal_df['protein_percent']=cereal_df['protein_cal']/cereal_df['calories']*100
```

```python
high_protein_df=cereal_df[cereal_df['protein_percent']>=20]
```

```python
high_protein_df.sort_values('protein_percent',ascending=False).head()
```

```python
mfr_df = cereal_df.groupby('mfr').mean(numeric_only=True)
```

```python
mfr_df.sort_values('sugar_per_cup',ascending=False)
```

```python
mfr_df['sugar_per_cup'].sort_values().plot(kind='bar',title='Average Sugar per Cup by Manufacturer',ylabel='Grams of Sugar',xlabel='Manufacturer',color='blue')
```

```python
shelf_df = cereal_df.groupby('shelf').mean(numeric_only=True)
```

```python
shelf_df.sort_values('sugar_per_cup', ascending=False).head()
```

```python
shelf_df['sugar_per_cup'].sort_values().plot(kind='bar',title='Average Suger Per Cup by Shelf Number',xlabel='Shelf',ylabel='Sugar per cup',color='purple')
```
