# Cereal Nutrition Analysis Project

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='darkgrid')
```

## Data Uploading and Cleaning

```python
cereal_df = pd.read_csv('cereal.csv')
cereal_df.head()
```

```python
cereal_df.describe()
```

## Exploratory Statistical Analysis

### Top 5 cereals with the most amount of sugar per serving 

```python
cereal_df.sort_values('sugars',ascending = False).head()
```

### Top 5 cereals with the most sugar per cup (using new variable sugar_per_cup)

```python
cereal_df['sugar_per_cup'] = cereal_df['sugars'] / cereal_df['cups']
```

```python
cereal_df.sort_values('sugar_per_cup', ascending = False).head(5)
```

### Top 5 cereals with the highest percent of calories from protein (using new variable protein_percent)

```python
cereal_df['protein_cal'] = cereal_df['protein'] * 4
```

```python
cereal_df['protein_percent'] = cereal_df['protein_cal'] / cereal_df['calories'] * 100
```

```python
high_protein_df = cereal_df[cereal_df['protein_percent'] >= 20]
```

```python
high_protein_df.sort_values('protein_percent', ascending = False).head()
```

### Highest amount of sugar per cup by cereal manufacturer

```python
mfr_df = cereal_df.groupby('mfr').mean(numeric_only=True)
```

```python
mfr_df.sort_values('sugar_per_cup', ascending = False)
```

## Visualizations and Takeaways

### Bar chart showing average sugar per cup based on manufacturer
### According to the chart, P (Post) had the highest average amount of sugar

```python
mfr_df['sugar_per_cup'].sort_values().plot(kind = 'bar', title = 'Average Sugar per Cup by Manufacturer', ylabel = 'Grams of Sugar', xlabel = 'Manufacturer', color = 'blue')
```

### Cereal is sorted by shelf number where it is displayed in stores to determine which shelf holds the cereals with the highest amount of sugar per cup 
### The bar chart shows that shelves 2 and 3 hold cereals with a higher number of grams of sugar per cup on average than shelf 1

```python
shelf_df = cereal_df.groupby('shelf').mean(numeric_only = True)
```

```python
shelf_df.sort_values('sugar_per_cup', ascending = False).head()
```

### Faulty data value (-1g of sugar) isolated

```python
negative = cereal_df[cereal_df['sugars'] == -1]
```

```python
negative
```

### Healthy cereals are defined as those produced by Kellogs (K) or General Mills (G) with less than 8g of sugar and at least 3g of fiber
### They are sorted below in a table and bar chart based on fiber content

```python
healthy_kids_df = cereal_df[(cereal_df['mfr'].isin (['K', 'G'])) & (cereal_df['sugars'] < 8) & (cereal_df['fiber'] >= 3)]
```

```python
healthy_kids_df.sort_values('fiber', ascending = False).head()
```

```python
healthy_kids_df['fiber'].sort_values().plot(kind = 'bar', title = 'Healthiest cereal', ylabel = 'Grams of fiber', color = 'blue')
```

```python
sns.barplot(cereal_df, x = 'shelf', y = 'sugar_per_cup')
plt.title('Average Sugar Per Cup for Each Shelf', fontsize = '14', weight = 'bold')
plt.xlabel('Retail Shelf Number')
plt.ylabel('Average Sugar Per Cup')
plt.savefig('sugar_per_shelf.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

### Finding: Sugar vs. Ratings
### The graph below shows the impact of grams per sugar per serving on a product's consumer rating score. It shows a noticeable negative correlation between the two, such that as grams of sugar increases, consumer rating score, in general, decreases. 

```python
sns.scatterplot(data = cereal_df, x = 'sugars', y = 'rating', hue = 'mfr', palette = 'muted', s = 50, alpha = 0.8)
plt.title("Impact of Sugar on Consumer Ratings", fontsize = 14,weight = 'bold')
plt.xlabel("Grams of Sugar per Serving")
plt.ylabel("Consumer Rating Score")
plt.savefig('impact_of_sugar_on_ratings.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

### Determining whether a relationship exists between grams of sugar and grams of carbohydrates
### No correlation between the two; cereals exist in all ranges, and a cereal higher in complex carbs does not necessarily have lower sugar

```python
sns.scatterplot(cereal_df, x = 'carbo', y = 'sugars', hue = 'mfr', palette = 'muted')
plt.title("Relationship Between Sugar and Carbohydrates", fontsize = 12, weight = 'bold')
plt.xlabel("Grams of Carbohydrates per Serving")
plt.ylabel("Grams of Sugar per serving")
plt.savefig('sugar_and_carbs.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

### Histogram showing the number of cereals with varying numbers of calories per serving 

```python
sns.histplot(data = cereal_df, x = 'calories', bins = 15, kde = True, color = 'teal')
plt.title("Full Market Distribution: Calories per Serving", fontsize = 14, weight = 'bold')
plt.xlabel("Calories")
plt.ylabel("Number of Cereals")
```

### Box Plot showing average amounts of fiber content for cereals on each retail shelf (1-3) as well as outliers
### Cereals on shelves 1 and 3 have more fiber on average than shelf 2, with shelf 3 having more higher fiber outliers

```python
plt.figure(figsize = (9,5))
sns.boxplot(data = cereal_df, x = 'shelf', y = 'fiber')
plt.title("Fiber Content by Retail Shelf", fontsize = 14, weight = 'bold')
plt.xlabel("Shelf Placement (1 = Bottom, 2 = Middle, 3 = Top)")
plt.ylabel("Grams of Fiber per Serving")
plt.savefig('fiber_per_shelf.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

### Correlations found and displayed on a heat map for numerical nutritional values
### Significant correlational values include a strong positive correlation between potassium and fiber, and significant negative correlations between both calories and ratings and sugars and ratings.

```python
all_numeric = cereal_df.select_dtypes(include = 'number')
```

```python
numeric_df = all_numeric.drop(columns = ['sugar_per_cup', 'protein_cal', 'protein_percent', 'cups', 'weight', 'cups', 'shelf', 'weight'])
```

```python
numeric_df
```

```python
numeric_df.corr()
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(numeric_df.corr(), annot = True, cmap = 'coolwarm', fmt = ".2f")
plt.title("Variable Correlation for Top 80 Cereals", fontsize = '14', weight = 'bold')
plt.savefig('variable_correlation.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
